```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = ['Branch','City','Customer Type','Gender','Payment']
```

```python
data = {'Customer type': ['Member', 'Normal']}
df = pd.DataFrame(data)

```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded_data = encoder.fit_transform(df[['Customer type']])
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out())
```

```python
print(encoded_df)
```

```python
# Group by 'Branch' and calculate the total sales for each branch
grouped_df = df.groupby('Branch')['Total'].sum()
print(grouped_df)
```

```python
# Group by 'Branch' and 'Product line', calculate total sales and average quantity
grouped_df = df.groupby(['Branch', 'Product line']).agg({'Total': 'sum', 'Quantity': 'mean'})
print(grouped_df)
```

```python
# Group by 'Branch' and 'Product line', calculate total sales and average quantity
grouped_df = df.groupby(['Branch', 'Product line']).agg({'Total': 'sum', 'Quantity': 'mean'})
print(grouped_df)
```

```python
def calculate_profit_margin(group):
    return (group['gross income'] - group['cogs']) / group['gross income'] * 100

# Group by 'Branch' and calculate profit margin for each branch
grouped_df = df.groupby('Branch').apply(calculate_profit_margin)
print(grouped_df)
```
